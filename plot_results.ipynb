Plotting results

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

infile = "sim_inputs_and_outputs/dynamicsCalcs.csv"

data = pd.read_csv(infile)
```

```python
plt.plot(data.loc[:,'r0'], data.loc[:,'Cell Net Q'])
plt.title("Net Rate of Heat Generation vs Distance")
plt.xlabel("Distance (m)")
plt.ylabel("Rate of Heat Generation (W)")
plt.grid()
plt.show()
```

```python
plt.plot(data.loc[:,'r0'], data.loc[:,'Cell Qgen'])
plt.title("Rate of Heat Generated by Battery vs Distance")
plt.xlabel("Distance (m)")
plt.ylabel("Rate of Heat Generation (W)")
plt.grid()
plt.show()
```

```python
plt.plot(data.loc[:,"t0"], data.loc[:,"T_m"], label="Motor Torque")

plt.legend
plt.title("Motor Torque vs Time")
plt.xlabel("Time (s)")
plt.ylabel("Torque (Nm)")
plt.grid()
plt.show()
```

Comparing Acceleration Runs with Peak and Continuous Torque Profiles

```python
import matplotlib.pyplot as plt
import pandas as pd
import json

infile = "sim_inputs_and_outputs/dynamicsCalcs_accel_peak.csv"

peak_df = pd.read_csv(infile)

infile = "sim_inputs_and_outputs/dynamicsCalcs_accel_cont.csv"

cont_df = pd.read_csv(infile)

infile = "data_prep_scripts/AMK_data.json"

with open(infile, "r") as file:
    AMK_data = json.load(file)

AMK_Peak = pd.DataFrame(AMK_data["PeakTorqueSpeed"])
AMK_Cont = pd.DataFrame(AMK_data["ContTorqueSpeed"])
```

Compare Original Data Torque Profiles

```python
# Plot 1
plt.plot(AMK_Peak.loc[:,"Speed"], AMK_Peak.loc[:,"Torque"], label="Peak")
plt.plot(AMK_Cont.loc[:,"Speed"], AMK_Cont.loc[:,"Torque"], label="Cont")

plt.grid()
plt.legend()
plt.title("Torque Speed (Cont vs Peak) from AMK Data")
plt.xlabel("Speed (rpm)")
plt.ylabel("Torque (Nm)")
plt.show()
```

```python
plt.plot(peak_df.loc[:,'t0'], peak_df.loc[:,'T_m'], label="Peak")
plt.plot(cont_df.loc[:,'t0'], cont_df.loc[:,'T_m'], label="Cont")
plt.legend()
```
