Determining maximum speed around track

```python
infile = "Sim_Autocross.csv"
inData = pd.read_csv(infile)

b_vec = (inData['Radius'] == 0) * 100000       # Create boolean vector for input data - and update the radius to be VERY large on straights
inData['Radius'] = inData.loc[:,'Radius'] + b_vec   # Then we update the radius vector

# And now... add a column to the dataframe with the maximum speed in each section (this is kinda like doing vector operations in MATLAB)
inData = inData.assign(MaxVelocity = np.sqrt(a_centrip * inData['Radius']))

# print(inData)
```
