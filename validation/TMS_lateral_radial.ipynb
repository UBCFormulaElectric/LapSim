Compare lateral vs radial cooling for our cells based on research done by Yu et al.

```python
import matplotlib.pyplot as plt
import numpy as np

R = 18/1000                         # m
H = 65/1000                         # m
kz = np.linspace(11.8, 15.4, 5)     # W/mK - from Yu et al. Lower average estimate
kr = 0.43                           # W/mK - from Bhundiya et al.

# Here to calculate dTr/dTz = kz*R^2 / 2*kr*H^2
this_kz_kr = kz / kr
this_dTr_dTz = this_kz_kr * (R**2) / (2 * H**2)

# Plot results
# y-axis: dTr/dTz
# x-axis: kz/kr
# One line at dTr/dTz = 1 (above the line, axial is better // below the line, radial is better)

num = 400
x = np.linspace(0, 150, num)
y_18650 = x * R**2 / (2 * H**2)
line = np.ones(num)
fill_x = [min(x), min(x), max(x), max(x)]
fill_y = [0, 1, 1, 0]
radial_x = 100
radial_y = 0.5
axial_x = 15
axial_y = 2

# Plot
plt.plot(x, y_18650)
plt.plot(x, line, 'k--')
plt.scatter(this_kz_kr, this_dTr_dTz, s=100, marker='*', edgecolors='k', label=('kz: ' + str(kz)))
plt.fill(fill_x, fill_y, "b", alpha=0.3)
plt.text(radial_x, radial_y, "Radial Cooling")
plt.text(axial_x, axial_y, "Axial Cooling")
plt.title("Radial vs Axial Heat Transfer Comparison for 18650s")
plt.xlabel("kz/kr")
plt.ylabel("dTr/dTz")
plt.legend()
plt.savefig("Radial_vs_Axial.png")
```
